```python
from skimage import data, io, util, segmentation, color, draw, filters
import numpy as np
from scipy import ndimage
%matplotlib inline
import matplotlib
from matplotlib import cm
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
from PIL import Image
import os 
import sys
import cv2
import json
from PIL import Image
```

```python
# find the list of all images
# cycle through them 
# for each image, ask to manually enter the FOV
# store all results in a JSON file
```

```python
path = "/Users/sfiligoj/Desktop/SEM_all"
os.chdir(path)
print(f"I'm now in folder {os.getcwd()} and will look into {path}")
filess = [f for f in os.listdir(path) if f.endswith('tif')]

results = dict.fromkeys(filess) # initialize the dictionary
```

```python
for file in filess:
    fig, ax = plt.subplots(figsize=(7,4))
    img = plt.imread(file)
    img = np.copy(img[1024:1100,500:600])
    ax.imshow(img)
    plt.show()
    print(f'file name:{file}')
    results[file] = int(input('Insert the FOV in nm: '))
    plt.close('all')
```

```python
with open('FOVs.json', 'w') as f:
    json.dump(results, f, indent=3)
with open('FOVs.json', 'r') as f:
    FOV_dict = json.loads(f.read())
```
